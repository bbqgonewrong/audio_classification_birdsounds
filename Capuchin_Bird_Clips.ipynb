##Function to load the audio data and resample them to 16kHz 
1. Data is loaded from the file specified
2. Data is converted to a wav tensor with 1 channel for mono audio
3. The tensor is squeezed in order to remove the trailing axis signifying a channel
4. Sample rate is made into a 64 bit integer
5. Audio is resampled from the original to 16khz


```python
def load_wav_16k_mono(filename):
  file_contents = tf.io.read_file(filename)
  wav,sample_rate = tf.audio.decode_wav(file_contents,desired_channels=1)
  wav = tf.squeeze(wav,axis = -1)
  sample_rate = tf.cast(sample_rate,dtype=tf.int64)
  wav = tfio.audio.resample(wav,rate_in=sample_rate,rate_out=16000)
  return wav
```
